```python
width = 20
window_width = 4

arr1 = np.linspace(1, width*2, width*2).astype(int)
print(arr1)

separate()
arr2 = arr1.reshape(2,width)
print(arr2)

separate()
sum = np.sum(arr2, axis=0)
plt.plot(sum, label='sum')

separate()
window = np.ones(window_width)
convolved = np.convolve(window, sum)
plt.plot(convolved, label='convolved')
plt.legend()
```
