```python
width = 20
window_width = 4

arr1 = np.linspace(1, width*2, width*2).astype(int)
print(arr1)

separate()
arr2 = arr1.reshape(2,width)
print(arr2)

separate()
sum = np.sum(arr2, axis=0)
plt.plot(sum, label='sum')

separate()
window = np.ones(window_width)
convolved = np.convolve(window, sum)
plt.plot(convolved, label='convolved')
plt.legend()
```

```python
img = cv2.imread('new-car-image.jpg')
[b, g, r] = cv2.split(img)

img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)

plt.imshow(img)

plt.figure()
plt.imshow(b, cmap='gray')

plt.figure()
plt.imshow(g, cmap='gray')

plt.figure()
plt.imshow(r, cmap='gray')
```

```python
ploty = np.linspace(1, 300, num=300)
np.random.seed(0)
plotx = np.array([y + 20*(y**2)*3e-4 + np.random.randint(-50, high=51) for y in ploty])
plt.plot(plotx, ploty, 'r.')

fitted = np.polyfit(ploty, plotx, 2)
a0, a1, a2 = fitted

curve = np.zeros_like(ploty)
for x in range(1, 300):
    y = a0*x**2 + a1*x + a2
    curve[x] = y
    
plt.plot(curve, ploty, 'b')
```
